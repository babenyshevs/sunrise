# 03 — Naive Model (Biased Baseline)

This notebook computes the raw churn difference between booster and non-booster users.
It is intentionally descriptive and **not causal**.

```python
import pandas as pd
from pathlib import Path

df = pd.read_csv(Path("processed/df_analysis.csv"))
print(f"Loaded {len(df):,} rows")
```

```python
naive = (
    df.groupby("has_booster", as_index=False)["churned"]
      .mean()
      .rename(columns={"churned": "churn_rate"})
)
naive["group"] = naive["has_booster"].map({0: "No Booster", 1: "Has Booster"})
naive["churn_rate_pct"] = naive["churn_rate"] * 100

p1 = float(naive.loc[naive["has_booster"] == 1, "churn_rate"].iloc[0])
p0 = float(naive.loc[naive["has_booster"] == 0, "churn_rate"].iloc[0])
delta_pp = (p1 - p0) * 100

display(naive[["group", "churn_rate_pct"]])
print(f"Naive churn gap (Booster - No Booster): {delta_pp:+.2f} pp")
print("Warning: this estimate is biased by self-selection and confounding.")
```
